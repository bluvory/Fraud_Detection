```python
import pandas as pd

df = pd.read_csv('../creditcard.csv')
df.head()
```

```python
from sklearn.cluster import KMeans

data = df.drop('Class', axis=1)
model = KMeans(n_clusters=3).fit(data)
```

```python
model.labels_
```

```python
data['cluster_id'] = model.labels_
```

```python
data
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
printcipalComponents = pca.fit_transform(data.drop('cluster_id', axis=1))
principalDf = pd.DataFrame(data=printcipalComponents, columns = ['principal component1', 'principal component2'])
```

```python
new = pd.concat([principalDf, data['cluster_id']], axis=1)
new.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot('principal component1', 'principal component2',
           data=new, fit_reg=False, scatter_kws={'s':150}, hue='cluster_id')

plt.title('kmean plot')
```

```python
pca = PCA(n_components=2)
printcipalComponents = pca.fit_transform(df.drop('Class', axis=1))
originalpc = pd.DataFrame(data=printcipalComponents, columns = ['original pca 1', 'original pca 2'])
```

```python
new2 = pd.concat([originalpc, df['Class']], axis=1)
new2.head()
```

```python
sns.lmplot('original pca 1', 'original pca 2',
           data=new2, fit_reg=False, scatter_kws={'s':150}, hue='Class')

plt.title('kmean plot')
```
